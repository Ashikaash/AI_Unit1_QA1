```python
from collections import deque
import heapq

initial_state = ([3,2,1], [], [])
goal_state = ([], [], [3,2,1])

# Utility to generate successors
def successors(state):
    res = []
    for i in range(3):
        if not state[i]:
            continue
        disk = state[i][-1]
        for j in range(3):
            if i == j:
                continue
            if state[j] and state[j][-1] < disk:
                continue
            new_state = [list(peg) for peg in state]
            new_state[i].pop()
            new_state[j].append(disk)
            res.append((tuple(tuple(peg) for peg in new_state), (i, j)))
    return res

# BFS

def bfs():
    queue = deque([(tuple(tuple(peg) for peg in initial_state), [])])
    visited = set()
    while queue:
        state, path = queue.popleft()
        if state == tuple(tuple(peg) for peg in goal_state):
            return path
        if state in visited:
            continue
        visited.add(state)
        for nxt, move in successors(state):
            queue.append((nxt, path + [move]))

if __name__ == "__main__":
    print("BFS:")
    print(bfs())
```
